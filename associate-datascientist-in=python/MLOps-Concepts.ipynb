Deployment strategies
While deploying the machine learning model, it is strongly recommended to test if the new model works as expected. To do this, you can use different deployment strategies.

In this exercise, you will see different characteristics of deployment strategies (basic, shadow, and canary deployment). Your task is to match each characteristic with the appropriate deployment strategy.

Certainly! Here are the characteristics and their matching deployment strategies:

Basic Deployment

Characteristic: The new version of the model is deployed to all users at once, replacing the old version immediately.
Explanation: This strategy involves a straightforward deployment where the new model completely replaces the old one without any gradual rollout.
Shadow Deployment

Characteristic: The new version of the model runs alongside the old version in a non-production environment to test its performance without affecting users.
Explanation: In this strategy, the new model processes requests and generates predictions, but these results are not served to users. It allows for comparison with the old model in real-world scenarios without impacting the live system.
Canary Deployment

Characteristic: The new version of the model is deployed to a small subset of users first before a full rollout to all users.
Explanation: This strategy gradually rolls out the new model to a small segment of users to monitor its performance and identify any issues before a full deployment.
These strategies help ensure that new models are thoroughly tested and any potential issues are identified early before affecting all users.

Great work! Picking the right strategy depends on concerns you might have during model deployment and helps you tackle any potential problems before they affect your business.

Examples of drift
Data drift often does not impact your machine learning model. However, concept drift can impact the quality of your machine learning model, since the patterns it was trained on earlier have changed.

In this exercise, you will classify which example belongs to which of the two, data drift or concept drift.

Congratulations! Being able to detect what type of drift your data might be suffering from can help to identify whether your model needs retraining. In this way, you can maintain a high-quality machine learning model

Here are examples of drift:

Data Drift:

Example: A retail company’s model was trained on historical sales data, but recently, the customer demographics have shifted, leading to different buying patterns that the model did not encounter before. This can result in discrepancies between the model’s predictions and the current data.
Concept Drift:

Example: A spam detection model was trained to identify spam emails based on patterns from a few years ago. Over time, spammers have changed their tactics and email content, making the model less effective at detecting new types of spam that weren't present in the original training data.
Concept Drift is particularly impactful because it means the underlying patterns or relationships that the model learned from are no longer valid, requiring adjustments or retraining to maintain model accuracy.

ChatGPT
Certainly! Here are the characteristics and their matching deployment strategies:

Basic Deployment

Characteristic: The new version of the model is deployed to all users at once, replacing the old version immediately.
Explanation: This strategy involves a straightforward deployment where the new model completely replaces the old one without any gradual rollout.
Shadow Deployment

Characteristic: The new version of the model runs alongside the old version in a non-production environment to test its performance without affecting users.
Explanation: In this strategy, the new model processes requests and generates predictions, but these results are not served to users. It allows for comparison with the old model in real-world scenarios without impacting the live system.
Canary Deployment

Characteristic: The new version of the model is deployed to a small subset of users first before a full rollout to all users.
Explanation: This strategy gradually rolls out the new model to a small segment of users to monitor its performance and identify any issues before a full deployment.
These strategies help ensure that new models are thoroughly tested and any potential issues are identified early before affecting all users.

Examples of drift
Data drift often does not impact your machine learning model. However, concept drift can impact the quality of your machine learning model, since the patterns it was trained on earlier have changed.
ChatGPT
Here are examples of drift:

Data Drift:

Example: A retail company’s model was trained on historical sales data, but recently, the customer demographics have shifted, leading to different buying patterns that the model did not encounter before. This can result in discrepancies between the model’s predictions and the current data.
Concept Drift:

Example: A spam detection model was trained to identify spam emails based on patterns from a few years ago. Over time, spammers have changed their tactics and email content, making the model less effective at detecting new types of spam that weren't present in the original training data.
Concept Drift is particularly impactful because it means the underlying patterns or relationships that the model learned from are no longer valid, requiring adjustments or retraining to maintain model accuracy.
